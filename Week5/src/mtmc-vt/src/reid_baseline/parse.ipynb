```python
import os
import sys
import cv2
import pickle
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image
```

```python
fs = open('feature.txt','r').readlines()
```

```python
fs = fs[1:]
```

```python
len(fs)
```

```python
dis = np.zeros((len(fs), len(fs)))
```

```python
m = {}

for l in fs:
    ls = l.split(' ')
    f_name = ls[0]
    fea = np.array([float(n) for n in ls[1:]])
    print(f_name, fea.shape)
    m[f_name.split('/')[-1]] = fea
```

```python
ks = list(m.keys())
```

```python
kmp = {}
for l in fs:
    ls = l.split(' ')
    f_name = ls[0]
    kmp[f_name.split('/')[-1]] = f_name
```

```python
fs[0]
```

```python
for i,k1 in enumerate(ks):
    if i %50 ==0:print(i)
    for j,k2 in enumerate(ks):
        f1 = m[k1]
        f2 = m[k2]
        dis[i,j] = np.linalg.norm(f1-f2)
```

```python
def plot_images(imgs, rows=5):
    figure = plt.figure(figsize=(13, 8))
    cols = len(imgs) // rows + 1

    for i in range(len(imgs)):
        subplot = figure.add_subplot(rows, cols, i + 1)
        subplot.axis('Off')
        plt.imshow(imgs[i], cmap='gray')
def plot_images_for_filenames(filenames, rows=5):
    imgs = [plt.imread(filename) for filename in filenames]
    
    return plot_images(imgs, rows)
```
